```python
import pandas as pd
```

```python
df = pd.DataFrame([[1,2,3],[4,5,6], [7,8,9]], columns = ["A", "B", "C"], index= ["D", "E", "F"])
```

df.head() shows first 5 rows by default, can also take arguments

```python
df.head()
```

We can see giving 2 as an argument to head() function now displays 2 rows only.

```python
df.head(2)
```

df.tail() shows bottom 5 rows by default and can also take arguments.

```python
df.tail()
```

Giving tail() an argument of 1 shows the bottom row of the data or array set.

```python
df.tail(1)
```

df.columns shows the column names as specified or just [1,2,3] by default.

```python
df.columns
```

df directly displays the whole array at once.

```python
df
```

df.index shows the row namings or [1,2,3] by default.

```python
df.index
```

Displays the index as a list

```python
df.index.tolist()
```

Displays columns as a list

```python
df.columns.tolist()
```

Displays information about all the rows

```python
df.info()
```
